```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#load the dataset
df = pd.read_csv("HousePricePrediction.csv")
# display first few rows of dataset
print(df.head())
```

```python
# Visualize the distribution of SalePrice
plt.figure(figsize=(10,6))
sns.histplot(df['SalePrice'],bins=30,kde=True)
plt.title('Distribution of SalePrice')
plt.xlabel('SalePrice')
plt.show()
```

```python
# Visualize correlation matrix
# Assuming 'df' is your DataFrame
# Drop non-numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Create the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Get the top correlated features with SalePrice
top_corr_features = correlation_matrix['SalePrice'].sort_values(ascending=False).index[1:4]
print("Top correlated features",top_corr_features)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

# Select features and target variable
features = df[top_corr_features]
target = df['SalePrice']
```

```python
# Check for missing values in the dataset
print("Missing values in the dataset:\n", df.isnull().sum())
```

```python
# Drop rows with missing values
df = df.dropna()

# Select features and target variable
features = df[top_corr_features]
target = df['SalePrice']

# Split the data into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size = 0.2,random_state = 42)
print('X_train',X_train)
print('X_test',X_test)
print('y_train',y_train)
print('y_test',y_test)
```

```python
# Build the Linear Regression model
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
print('y_pred',y_pred)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Mean Square Error: {mse}")
print(f"R-squared: {r2}")
```
